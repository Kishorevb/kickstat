```python
import json
import csv
import os
import pandas as pd
from flatten_json import flatten
from pathlib import Path
```

```python
rootdir = 'C:\\Users\\krish.vanapalli\\Documents\\Work\\KickStat\\open-data-master\\data\\lineups\\'
```

```python
dataframes_supported_entities = {}
dataframes_supported_entities["player"] = pd.DataFrame()
dataframes_supported_entities["country"] = pd.DataFrame()
dataframes_supported_entities["player_team"] = pd.DataFrame()
dataframes_supported_entities["player_country"] = pd.DataFrame()
dataframes_supported_entities["match_player"] = pd.DataFrame()
dataframes_supported_entities["match_player_cards"] = pd.DataFrame()
dataframes_supported_entities["match_player_positions"] = pd.DataFrame()
```

```python
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        match_id = Path(file).stem
        print(match_id)
        with open(os.path.join(subdir, file), encoding='utf-8') as f:
            teams_list = json.load(f)
        for team in teams_list:
            team_id = team["team_id"]
            for player in team["lineup"]:
                player_dict = {"player_id": player["player_id"], "player_name": player["player_name"], "player_nickname": player["player_nickname"]}
                df_dictionary_tmp = pd.DataFrame([player_dict])
                dataframes_supported_entities["player"] = pd.concat([dataframes_supported_entities["player"], df_dictionary_tmp], ignore_index=True)                                                                     
                
                player_team_dict = {"player_id": player["player_id"], "team_id": team_id}
                df_dictionary_tmp = pd.DataFrame([player_team_dict])
                dataframes_supported_entities["player_team"] = pd.concat([dataframes_supported_entities["player_team"], df_dictionary_tmp], ignore_index=True)                                                                     
                
                if "country" in player:
                    country_dict = {"id": player["country"]["id"], "name": player["country"]["name"] }
                else:
                    country_dict = {"id": None, "name": None }
                df_dictionary_tmp = pd.DataFrame([country_dict])
                dataframes_supported_entities["country"] = pd.concat([dataframes_supported_entities["country"], df_dictionary_tmp], ignore_index=True)                                                                     
                                                                                        
                if "country" in player:
                    player_country_dict = {"player_id": player["player_id"], "country_id": player["country"]["id"]}
                else:
                    player_country_dict = {"player_id": player["player_id"], "country_id": None}
                df_dictionary_tmp = pd.DataFrame([player_country_dict])
                dataframes_supported_entities["player_country"] = pd.concat([dataframes_supported_entities["player_country"], df_dictionary_tmp], ignore_index=True)                                                                     
                                                                                         
                match_player_dict = {"match_id": match_id, "player_id": player["player_id"], "jersey_number": player["jersey_number"]}
                df_dictionary_tmp = pd.DataFrame([match_player_dict])
                dataframes_supported_entities["match_player"] = pd.concat([dataframes_supported_entities["match_player"], df_dictionary_tmp], ignore_index=True)                                                                     
                                                                                         
                for card in player["cards"]:
                    match_player_cards = {"match_id": match_id, "player_id": player["player_id"], "time": card["time"], "card_type": card["card_type"], "reason": card["reason"], "period": card["period"]}
                    df_dictionary_tmp = pd.DataFrame([match_player_cards])
                    dataframes_supported_entities["match_player_cards"] = pd.concat([dataframes_supported_entities["match_player_cards"], df_dictionary_tmp], ignore_index=True)
                for position in player["positions"]:
                    match_player_positions = {"match_id": match_id, "player_id": player["player_id"], "position_id": position["position_id"], "position": position["position"], "from": position["from"], "to": position["to"], "from_period": position["from_period"], "to_period": position["to_period"], "start_reason": position["start_reason"], "end_reason": position["end_reason"]}
                    df_dictionary_tmp = pd.DataFrame([match_player_positions])
                    dataframes_supported_entities["match_player_positions"] = pd.concat([dataframes_supported_entities["match_player_positions"], df_dictionary_tmp], ignore_index=True)                                                                     
                                                                                         
```

```python
for key, value_df in dataframes_supported_entities.items():
    value_df = value_df.loc[value_df.astype(str).drop_duplicates().index]
    output_file_name = f"C:\\Users\\krish.vanapalli\\Documents\\Work\\KickStat\\KBDataFiles\\LineUps\\{key}.csv"
    value_df.to_csv(output_file_name, sep=',', encoding='utf-8', index=False, header=True)      
```
